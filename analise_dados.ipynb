###MODULO 4.6 - CORRELACAO


````python
correlacao_continua = dados['IDADE'].corr(dados['SALARIO'])
correlacao_continua
#CORRELACAO - ajuda entender se as variaveis se movem juntas ou nao
# resultado 0.29 ta mais perto do 0, apesar de ser positiva, nao é uma correlacao forte entre as variaveis
# se tiver mais perto do 1, é mais positiva.
# qto maior a idade n necessariamente maior o salario
````

````python
from scipy.stats import (chi2_contingency)
````

````python
def cramer_coeficiente(coluna1, coluna2):
  tabela_cruzada = np.array(pd.crosstab(coluna1, coluna2))
  chi2 = chi2_contingency(tabela_cruzada)[0]
  soma = np.sum(tabela_cruzada)
  mini = min(tabela_cruzada.shape) -1
  cramer = np.sqrt(chi2/(soma*mini))
  return cramer
  #esse 0 retorna apenas o primeiro valor numerico
````

````python
cramer_coeficiente(dados['COR/RACA/ETNIA'], dados['NIVEL DE ENSINO'])
# com esse resultado indica que as colunas nao tem uma correlacao
````

````python
cramer_coeficiente(dados['GENERO'], dados['NIVEL DE ENSINO'])
# RESULTADO INDICA QUE N HA RELACAO POSITIVA ENTRE GENERO E NIVEL DE ENSINO.
````
